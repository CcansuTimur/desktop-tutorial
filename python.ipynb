**if, for ve Fonksiyonların Birlikte Kullanımı**

```python
#mini uygulama: maaşı 3000 tl'den fazla olanlara %10, 3000 tl'den az olanlara %20 zam yapalım.

maaşlar = [1000,2000,3000,4000,5000]

def maaş_üst(x):
    print(x*10/100 + x)
    
def maaş_alt(x):
    print(x*20/100 + x)
    
for i in maaşlar:
    if i >= 3000:
        maaş_üst(i)
    else:
        maaş_alt(i)
    
```

* **Filter**

```python
df = pd.DataFrame({"gruplar": ["A", "B", "C", "A", "B", "C"],
                   "değişken1": [10, 23, 33, 22, 11, 99],
                   "değişken2": [100, 253, 333, 262, 111, 969]})
df
```

```python
def filter_func(x):
    return x["değişken1"].std() > 9
```

```python
df.groupby("gruplar").std()
```

```python
 df.groupby("gruplar").filter(filter_func)
```
